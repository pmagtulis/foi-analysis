```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
df = pd.concat(
    map(pd.read_csv, ['output/foi1.csv', 'output/foi2.csv', 'output/foi3.csv', 'output/foi4.csv', 'output/foi5.csv', 'output/foi6.csv', 'output/foi7.csv', 'output/foi8.csv', 'output/foi9.csv', 'output/foi10.csv', 'output/foi11.csv', 'output/foi12.csv', 'output/foi13.csv', 'output/foi14.csv', 'output/foi15.csv', 'output/foi16.csv', 'output/foi17.csv', 'output/foi18.csv', 'output/foi19.csv', 'output/foi20.csv' ]), ignore_index=True).drop_duplicates()
df
```

```python
df.date = pd.to_datetime(df.date)
df
```

```python
df2 = pd.read_csv ('foi_final.csv')
df2 = df2.drop(['Unnamed: 0'], axis=1)
df2
```

```python
df_final = pd.concat([df, df2]).drop_duplicates().reset_index(drop=True)
df_final
```

```python
df_final.date = pd.to_datetime(df_final.date)
```

```python
df_final = df_final.drop_duplicates().sort_values('date').reset_index(drop=True)
df_final
```
